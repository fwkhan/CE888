```python
data = country_data.drop('country',axis=1)
```

```python
# Fit K-Means Clustering on noise moon data
km=cluster.KMeans(n_clusters=3)
km.fit(data)
km.labels_
```

```python
new_data = country_data
new_data['label'] = km.labels_
```
